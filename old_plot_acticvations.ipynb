```python
def get_sigmoid_points():
    
    title = "Sigma Function"
    label = ""

    x_points = np.linspace(-10, 10, 100)
    y_points = 1 / (1 +np.exp(-x_points))
    
    # Calculate limits...
    # Function as it approaches negative infinity
    lower_limit = 1 / (1 +np.exp(-(-np.inf)))
    # Function as it approaches positive infinity
    upper_limit = 1 / (1 +np.exp(-(np.inf)))
    
    # List significant values. 
    horizontal_lines = [lower_limit, 0.5, upper_limit]
    vertical_lines = [0.0]
    
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

```python
def get_tanh_points():
    
    title = "Hyperbolic tangent function (tanh)"
    label = ""

    x_points = np.linspace(-10, 10, 100)
    y_points = (np.exp(x_points) - np.exp(-x_points)) / (np.exp(x_points) + np.exp(-x_points))
    
    # Calculate limits... Inputting negative and positive infinity produces Nan
    #lower_limit = (np.exp(-np.inf) - np.exp(np.inf)) / (np.exp(-np.inf) + np.exp(np.inf))
    #upper_limit = (np.exp(np.inf) - np.exp(-np.inf)) / (np.exp(np.inf) + np.exp(-np.inf))
    
    # To avoid Nan, just set lower and upper limits manually to -1 and 1 respectively to be represented as horizontal lines.
    # List significant values. 
    horizontal_lines = [-1.0, 0.0, 1.0]
    vertical_lines = [0.0]
    
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

```python
def get_linear_points():
    
    title = "Linear Function"
    label = ""

    #Define points 
    x_points = np.linspace(-10, 10, 100)
    y_points = x_points
    
    # List significant values - (the x and y axes)
    horizontal_lines = [0.0]
    vertical_lines = [0.0]
    
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

```python
def get_relu_points():
    
    title = "Rectified Linear Unit Function (ReLU)"
    label = ""

    # Define points
    x_points = np.linspace(-10, 10, 100)
    
    # For y_points, replace the values less than zero with zero 
    # (that is, negative values are clipped to zero)
    y_points = np.clip(x_points, a_min= 0.0, a_max=None)

    # List significant values. 
    horizontal_lines = [0.0]
    vertical_lines = [0.0]
    
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

```python
def get_leaky_relu_points(alpha = 0.01):
    
    title = "Leaky Rectified Linear Unit Function (Leaky ReLU)"
    label = "alpha="+str(alpha)

    # Define points
    x_points = np.linspace(-10, 10, 100)
    # For y_points, if value is greater than  zero it remains the same,
    # otherwise it is multiplued by the alpha value
    y_points = np.where(x_points > 0, x_points, x_points*alpha)

    # List significant values. 
    horizontal_lines = [0.0]
    vertical_lines = [0.0]
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

```python
def get_elu_points(alpha = 0.5):
    
    title = "Exponential Linear Unit (ELU)"
    label = "alpha="+str(alpha)

    # Define points
    x_points = np.linspace(-10, 10, 100)
    # For y_points, if value is greater than  zero it remains the same,
    # otherwise it is calculated to be alpha * (e**x -1)
    y_points = np.where(x_points > 0, x_points, alpha*((np.exp(x_points)) -1))

    # List significant values. 
    horizontal_lines = [0.0]
    vertical_lines = [0.0]
    
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

### Define a function that graphs the individual functions with matplotlib

```python
def plot_function(x_points, y_points, title="", label="", horizontal_lines=[], vertical_lines=[], color="red", show_plot = True):
    
    # Plot the x and y points in red
    plt.plot(x_points, y_points, color=color, linewidth=2, label=label)
    
    # Add significant horizontal lines
    for line in horizontal_lines:
        plt.axhline(y=line, color="grey", linestyle=":")
    
    # Add significant vertical lines
    for line in vertical_lines:
        plt.axvline(x=line, color="grey", linestyle=":")
        
    # Limit the x axes to the first and last x_points
    plt.xlim(x_points[0], x_points[-1])
    
    # Label graph
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    
    # If the plot contains a label, add a legend
    if label:
        plt.legend()
    
    # Save to file 
    #filename = function_name + ".png"
    #plt.savefig(filename)
    
    # Show plot
    if show_plot:
        plt.show()
        
    
```

## Plotting multiple graphs

The plot_function() can plot several of the same type of graph but with varying alpha values.

### Multi-plot example 1, Leaky Rectified Linear Unit (Leaky ReLU) with different alpha values

```python
# Use the get_relu_points() function to get data for the the relu function plots
# For each call to "plot_function()" do the following:
# 1. Change the alpha value (default is 0.01)
# 2. Change the color.  
# 3. Also, "set_plot" to False for all plots except the last plot

x_points, y_points, function_name, label, horizontal_lines, vertical_lines = get_leaky_relu_points()
plot_function(x_points, y_points, function_name, label, horizontal_lines, vertical_lines, color="red", show_plot=False)

x_points, y_points, function_name, label, horizontal_lines, vertical_lines = get_leaky_relu_points(0.02)
plot_function(x_points, y_points, function_name, label, horizontal_lines, vertical_lines, color="orange", show_plot=False)

x_points, y_points, function_name, label, horizontal_lines, vertical_lines = get_leaky_relu_points(0.05)
plot_function(x_points, y_points, function_name, label, horizontal_lines, vertical_lines, color="green", show_plot=False)

x_points, y_points, function_name, label, horizontal_lines, vertical_lines = get_leaky_relu_points(0.10)
plot_function(x_points, y_points, function_name, label, horizontal_lines, vertical_lines, color="blue", show_plot=True)
```

## Import libraries

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [7,7]
```

## Define the functions

### Define individual functions that get the points for each activation function

```python
def get_points(activation_function_name, alpha= 0.01):
    
    #Define the x values, this will be the same for each plot
    start_x = -10
    stop_x = 10
    num_x_points = 100
    x_points = np.linspace(start_x, stop_x, num_x_points)
    
    # Define y values and significant x and y values for each function
    if activation_function_name == "linear": 
        title = "Linear Function"
        label = ""
        
        #Calculate y values
        y_points = x_points
    
        # List significant values - (the x and y axes)
        horizontal_lines = [0.0]
        vertical_lines = [0.0]
    
    elif activation_function_name == "relu":
        title = "Rectified Linear Unit (ReLU)"
        label = ""
    
        # For y_points, replace the values less than zero with zero 
        # (that is, negative values are clipped to zero)
        y_points = np.clip(x_points, a_min= 0.0, a_max=None)

        # List significant values. 
        horizontal_lines = [0.0]
        vertical_lines = [0.0]
        
    elif activation_function_name == "leaky_relu":
        title = "Leaky Rectified Linear Unit (Leaky ReLU)"
        label = "alpha="+str(alpha)

        # For y_points, if value is greater than  zero it remains the same,
        # otherwise it is multiplued by the alpha value
        y_points = np.where(x_points > 0, x_points, x_points*alpha)

        # List significant values. 
        horizontal_lines = [0.0]
        vertical_lines = [0.0]
        
    elif activation_function_name == "elu":
        title = "Exponential Linear Unit (ELU)"
        label = "alpha="+str(alpha)

        # For y_points, if value is greater than  zero it remains the same,
        # otherwise it is calculated to be alpha * (e**x -1)
        y_points = np.where(x_points > 0, x_points, alpha*((np.exp(x_points)) -1))

        # List significant values. 
        horizontal_lines = [0.0]
        vertical_lines = [0.0]
        
    elif activation_function_name == "sigmoid": 
        title = "Sigmoid Function"
        label = ""

        y_points = 1 / (1 +np.exp(-x_points))
    
        # Calculate limits...
        # Function as it approaches negative infinity
        lower_limit = 1 / (1 +np.exp(-(-np.inf)))
        # Function as it approaches positive infinity
        upper_limit = 1 / (1 +np.exp(-(np.inf)))
    
        # List significant values. 
        horizontal_lines = [lower_limit, 0.5, upper_limit]
        vertical_lines = [0.0]
    
    elif activation_function_name == "tanh":
        title = "Hyperbolic tangent function (tanh)"
        label = ""

        y_points = (np.exp(x_points) - np.exp(-x_points)) / (np.exp(x_points) + np.exp(-x_points))
    
        # Calculate limits... Inputting negative and positive infinity produces Nan
        #lower_limit = (np.exp(-np.inf) - np.exp(np.inf)) / (np.exp(-np.inf) + np.exp(np.inf))
        #upper_limit = (np.exp(np.inf) - np.exp(-np.inf)) / (np.exp(np.inf) + np.exp(-np.inf))
    
        # To avoid Nan, set lower and upper limits manually to -1 and 1 respectively to be represented as horizontal lines.
        # List significant values. 
        horizontal_lines = [-1.0, 0.0, 1.0]
        vertical_lines = [0.0]
        
    else:
        #If an activation name is not defined, (eg. action_function_name is mispelt) draw a red box with a cross in it
        title = activation_function_name + " is not defined."
        label = ""
        x_points = [start_x, start_x, stop_x, stop_x, start_x, start_x, stop_x, stop_x, stop_x, start_x, stop_x]
        y_points = [start_x, stop_x, stop_x, start_x, start_x, stop_x, start_x, stop_x, start_x, start_x, stop_x]
    
        horizontal_lines = []
        vertical_lines = []
        
    return x_points, y_points, title, label, horizontal_lines, vertical_lines
```

### Define a function that graphs the individual functions with matplotlib

```python
def plot_points(x_points, y_points, title="", label="", horizontal_lines=[], vertical_lines=[]):
    
    # Plot the x and y points in red
    plt.plot(x_points, y_points, linewidth=2, color="red", label=label)
    
    # Add significant horizontal lines
    for line in horizontal_lines:
        plt.axhline(y=line, color="grey", linestyle=":")
    
    # Add significant vertical lines
    for line in vertical_lines:
        plt.axvline(x=line, color="grey", linestyle=":")
        
    # Limit the x axes to the first and last x_points
    plt.xlim(x_points[0], x_points[-1])
    
    # Label graph
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    
    # If the plot contains a label, add a legend
    if label:
        plt.legend()
    
    # Save to file 
    #filename = function_name + ".png"
    #plt.savefig(filename)
    
    # Show plot
    plt.show()
        
    
```

## Use functions 

### Example 1:  Linear activation function

```python
# Define the type of activation function
activation_name = "linear"

# Use the get_sigmoid_points() function to get data for the the sigmoid function graph
x_points, y_points, title, label, horizontal_lines, vertical_lines = get_points(activation_name)

# Use the plot_points() function to make a plot
plot_points(x_points, y_points, title, label, horizontal_lines, vertical_lines)
```
